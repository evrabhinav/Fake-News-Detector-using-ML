```python
import nltk
nltk.download('stopwords')

import pandas as pd

# Reading the dataset from a CSV file named 'train.csv'


dataset = pd.read_csv(r'C:\Users\ADMN\Downloads\train (1).csv')

# Displaying the dataset
print(dataset)

# Displaying the list of English stopwords from NLTK
from nltk.corpus import stopwords
print(stopwords.words('english'))

# Displaying the first few rows of the dataset
print(dataset.head())

# Checking for missing values in the dataset and displaying their sums per column
print(dataset.isnull().sum())
```

```python
# Replacing null values with an empty string
dataset = dataset.fillna('')

# Merging authors' name and news title into the 'content' column
dataset['content'] = dataset['text'] + ' ' + dataset['title']
dataset['content']

# Separating the data and labels
X = dataset.drop(columns=['label'], axis=1)
Y = dataset['label']
print(X)
print(Y)

# Stemming - reducing a word to its root word using PorterStemmer
import string
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stop_word = stopwords.words('english')
porter_stem = PorterStemmer()

def cleaning_data(x):
    x = x.lower()
    x = ' '.join([word for word in x.split() if word not in stop_word])
    return x

dataset['text'] = dataset['text'].apply(cleaning_data)
```

```python
# Cleaning the 'content' column using the cleaning_data function
dataset['content'] = dataset['content'].apply(cleaning_data)
M = dataset['content']

# Separating data and labels
X = dataset['content'].values
Y = dataset['label'].values

# Converting textual data to numerical data using TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
X
Y

vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X = vectorizer.transform(X)
print(X)

# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initializing the Logistic Regression model
model = LogisticRegression()

# Fitting the model on the training data
model.fit(X_train, Y_train)

# Accuracy score on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy score of the training data:", training_data_accuracy)

# Accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy score of the test data:", test_data_accuracy)

# Making a prediction on new data (X_new)
X_new = X_test[3]
prediction = model.predict(X_new)
print(prediction)

# Checking the prediction and printing the result
if prediction[0] == 0:
    print('The news is Real')
else:
    print('The news is Fake')
```
